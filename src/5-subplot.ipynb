### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Data for plots
    x = np.linspace(0, 10, 100)
    y = x ** 2
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    scatter_x = [1, 2, 3, 4, 5]
    scatter_y = [2, 4, 6, 8, 10]
    pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
    pie_sizes = [40, 25, 20, 15]
    pie_colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    
    # Line plot
    axs[0, 0].plot(x, y)
    axs[0, 0].set_title('Line Plot')
    axs[0, 0].set_xlabel('x values')
    axs[0, 0].set_ylabel('y = x^2')

    # Bar plot
    axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'purple'])
    axs[0, 1].set_title('Bar Plot')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')

    # Scatter plot
    axs[1, 0].scatter(scatter_x, scatter_y)
    for i in range(len(scatter_x)):
        axs[1, 0].annotate(f'({scatter_x[i]}, {scatter_y[i]})', (scatter_x[i], scatter_y[i]), textcoords="offset points", xytext=(0,10), ha='center')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('x values')
    axs[1, 0].set_ylabel('y values')

    # Pie chart
    axs[1, 1].pie(pie_sizes, labels=pie_labels, colors=pie_colors, autopct='%1.1f%%', startangle=140)
    axs[1, 1].set_title('Pie Chart')

    # Adjust layout
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
